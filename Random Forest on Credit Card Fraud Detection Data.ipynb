```python
from sklearn.ensemble._hist_gradient_boosting.gradient_boosting import train_test_split
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
```

```python
creditcard_dataset = fetch_openml(name = 'creditcard', version = 1, as_frame = True)
```

```python
# Separate target variable
x = df.drop('Class', axis = 1)
y = df['Class'].astype(int)
```

```python
# Split dataset into train and test
x_train , x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42, stratify = y)
```

```python
# Create Random Forest, class_weight='bllanced helps model for giving importance fraude cases and
# n_jobs=-1 use all core of processor to train model faster
rf = RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced', n_jobs=-1)
```

```python
rf.fit(x_train, y_train)
```

```python
prediction = rf.predict(x_test)
```

```python
print(confusion_matrix(y_test, prediction))
```

```python
print(classification_report(y_test, prediction))
```

```python
# calculate only fraud
pred_proba = rf.predict_proba(x_test)[:,1]
```

```python
# Set new threshhold
custom_threshold = 0.3
pred_custom = (pred_proba >=custom_threshold).astype(int)
```

```python
print(confusion_matrix(y_test, pred_custom))
```

```python
print(classification_report(y_test, pred_custom))
```
